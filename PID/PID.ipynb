```python
import numpy as np
import particle_data as pd
import Logisitic_regression as lr
import matplotlib.pyplot as plt

np.set_printoptions(precision=3, suppress=False, threshold=40)
```

```python

N1 = 5000
N2 = 5000
N_it = 10000
m1 = 1.0
m2 = 1.1
X,Y = pd.generate_training_set(m1,m2,N1,N2)
X_rescale,means,stds = lr.rescale(X)

m,n = X.shape
w_in = np.zeros(n)
b_in=0.0

w,b,_,_ = lr.log_reg(X_rescale,Y,w_in,b_in,N_it,alpha=1.0,gamma=0)

X_new,Y_new = pd.generate_training_set(m1,m2,N1,N2)
X_new_rescale = (X_new-means)/stds

pred_rate = (np.sum((lr.prediction(X_new_rescale,w,b) == Y_new).astype(float))/(N1+N2))*100

print(f'Only linear terms : With masses {m1} and {m2}, the success rate is {pred_rate:.4}%\n')
print(f'Weights are w={w}\n')
print(f'Bias is b={b:.4}')
```

```python
plt.scatter(X[0:N1,0], Y[0:N1],color = 'blue')
plt.scatter(X[N1:-1,0], Y[N1:-1],color = 'red')
plt.xlabel("x")
plt.ylabel("y")
plt.xscale("log")
plt.title("Scatter plot")
plt.show()
```

```python
# X_quad = (np.concatenate([X,X**2],axis=1))
X_quad = X**2
# X_quad = X[:,0]**2-X[:,1]**2-X[:,2]**2-X[:,3]**2
X_quad_rescale,quad_means,quad_stds = lr.rescale(X_quad)

m,n = X_quad.shape
w_in = np.ones(n)
b_in=0.0

w,b,_,_ = lr.log_reg(X_quad_rescale,Y,w_in,b_in,N_it,alpha=1.0,gamma=0.0)

X_new,Y_new = pd.generate_training_set(m1,m2,N1,N2)
# X_new_quad = (np.concatenate([X_new,X_new**2],axis=1))
X_new_quad = X_new**2
# X_new_quad =X_new[:,0]**2-X_new[:,1]**2-X_new[:,2]**2-X_new[:,3]**2
X_new_quad_rescale = (X_new_quad-quad_means)/quad_stds

pred_rate = (np.sum((lr.prediction(X_new_quad_rescale,w,b) == Y_new).astype(float))/(N1+N2))*100

print(f'With quadartic terms : With masses {m1} and {m2}, the success rate is {pred_rate:.4}%')
print(f'Weights are w={w}\n')
print(f'Bias is b={b:.4}')
```
